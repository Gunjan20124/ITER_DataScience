```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]
plt.plot(x, y)   # simple line graph
plt.show()
```

```python
plt.scatter(x, y)
plt.show()
```

```python
categories = ['Austin', 'Bangkok', 'Catopia']
values = [3, 7, 5]
plt.bar(categories, values)
plt.show()
```

```python
data = [1,1,2,2,2,3,3,4,4,5,5,5,5,6,7,8]
plt.hist(data, bins=5)   
plt.show()
```

```python
sizes = [40, 30, 20, 10]
labels = ['A', 'B', 'C', 'D']
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.show()
```

```python
plt.plot(x, y, label="Line")
plt.title("Simple Plot")     
plt.xlabel("X-axis")         
plt.ylabel("Y-axis")         
plt.legend()                  
plt.show()
```

```python
plt.plot(x, y, color="red", marker="o", linestyle="--")
plt.show()
```

```python
plt.subplot(1,2,1)  
plt.plot([1,2,3],[1,4,9])

plt.subplot(1,2,2)
plt.plot([1,2,3],[1,2,3])
plt.show()
```
