```python
import pandas as pd
import numpy as np

df = pd.read_csv(r"C:\Users\tparida\Downloads\public_transport_delays.csv")

df["weather_condition"] = df["weather_condition"].fillna("Unknown")
df["destination_station"] = df["destination_station"].fillna("None")
df["event_type"] = df["event_type"].fillna("None")

df = df.dropna(subset=["actual_departure_delay_min", "actual_arrival_delay_min", "temperature_C"])
```

```python
print(df)
```

```python

print(df[["actual_departure_delay_min", "actual_arrival_delay_min", "temperature_C"]].describe())

# Mean, median, std
summary = {
    "Mean Departure Delay": np.mean(df["actual_departure_delay_min"]),
    "Median Departure Delay": np.median(df["actual_departure_delay_min"]),
    "Std Departure Delay": np.std(df["actual_departure_delay_min"]),
    "Mean Arrival Delay": np.mean(df["actual_arrival_delay_min"]),
    "Median Arrival Delay": np.median(df["actual_arrival_delay_min"]),
    "Std Arrival Delay": np.std(df["actual_arrival_delay_min"]),
    "Mean Temp (°C)": np.mean(df["temperature_C"]),
    "Avg Humidity (%)": np.mean(df["humidity_percent"]),
    "Avg Wind Speed (km/h)": np.mean(df["wind_speed_kmh"]),
}
summary_df = pd.DataFrame(summary, index=[0])
summary_df
```

```python

avg_delay_weather = df.groupby("weather_condition")[["actual_departure_delay_min", "actual_arrival_delay_min"]].mean()

avg_delay_event = df.groupby("event_type")[["actual_departure_delay_min", "actual_arrival_delay_min"]].mean()

avg_delay_peak = df.groupby("peak_hour")[["actual_departure_delay_min", "actual_arrival_delay_min"]].mean()
print(df)
```

```python

summary_df.to_csv("transport_delay_summary.csv", index=False)
avg_delay_weather.to_csv("avg_delay_by_weather.csv")
avg_delay_event.to_csv("avg_delay_by_event.csv")
avg_delay_peak.to_csv("avg_delay_by_peak.csv")

print(" exported successfully!")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10,6)

plt.figure()
plt.plot(df['time'], df['actual_departure_delay_min'], marker='o', color='blue')
plt.title("Actual Departure Delay over Time")
plt.xlabel("Time")
plt.ylabel("Departure Delay (minutes)")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

avg_delay_transport = df.groupby('transport_type')['actual_departure_delay_min'].mean()
avg_delay_transport.plot(kind='bar', color='orange')
plt.title("Average Departure Delay by Transport Type")
plt.xlabel("Transport Type")
plt.ylabel("Average Delay (minutes)")
plt.show()

plt.scatter(df['traffic_congestion_index'], df['actual_departure_delay_min'], c='green', alpha=0.7)
plt.title("Traffic Congestion vs Departure Delay")
plt.xlabel("Traffic Congestion Index")
plt.ylabel("Departure Delay (minutes)")
plt.show()
```

```python
delays = df["actual_departure_delay_min"].to_numpy()
print("Mean Departure Delay:", np.mean(delays))
print("Median Departure Delay:", np.median(delays))
print("Std Deviation:", np.std(delays))

plt.figure()
sns.boxplot(data=df, x='transport_type', y='actual_departure_delay_min', palette='Set2')
plt.title("Departure Delay by Transport Type")
plt.xlabel("Transport Type")
plt.ylabel("Departure Delay (minutes)")
plt.show()

plt.figure()
sns.barplot(data=df, x='peak_hour', y='actual_departure_delay_min', ci=None, palette='Set1')
plt.title("Average Departure Delay vs Peak Hour")
plt.xlabel("Peak Hour (0=Off-peak, 1=Peak)")
plt.ylabel("Average Departure Delay")
plt.show()

plt.figure()
sns.scatterplot(data=df, x='traffic_congestion_index', y='actual_departure_delay_min', hue='transport_type')
plt.title("Traffic Congestion vs Departure Delay")
plt.xlabel("Traffic Congestion Index")
plt.ylabel("Departure Delay (minutes)")
plt.show()

plt.figure()
sns.barplot(data=df, x='season', y='actual_departure_delay_min', ci=None, palette='Set3')
plt.title("Average Departure Delay by Season")
plt.xlabel("Season")
plt.ylabel("Average Delay")
plt.show()

numeric_df = df.select_dtypes(include=np.number)
plt.figure(figsize=(10,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Numeric Columns")
plt.show()

df.to_csv("public_transport_cleaned.csv", index=False)
print("Cleaned dataset saved successfully!")
```
